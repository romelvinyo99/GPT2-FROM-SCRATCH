## ------------------------Simplified-Attention Mechanism-------------------------------

## Token embeddings - Capture semantic meaning

```python
# Words / Tokens since we will assume the word-based tokenizer for simplicity

corresponding_words = ["Your", "journey", "begins", "with", "one", "step"]
corresponding_words
```

```python
# Creating the token embeddings - Randomized
output_dim = 3

import random
random.seed(42)
inputs = torch.tensor([
    [0.43, 0.15, 0.89], # Your    # X1
    [0.55, 0.87, 0.66], # journey # x2
    [0.57, 0.85, 0.64], # begins  # X3
    [0.22, 0.58, 0.33], # with    # X4
    [0.77, 0.25, 0.10], # one     # X5
    [0.05, 0.80, 0.55] # step     # X6
])  
inputs
```

```python
# 3D plot of word embeddings

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Extracting x, y, z coordinates from the token embeddinds
x_coords = inputs[:, 0].numpy()
y_coords = inputs[:, 1].numpy()
z_coords = inputs[:, 2].numpy()

# Creating 3d plots
fig = plt.figure(figsize=(6,10))
ax = fig.add_subplot(111, projection="3d")

for x, y, z, word in zip(x_coords, y_coords, z_coords, corresponding_words):
    ax.scatter(x, y, z)
    ax.text(x, y, z, word, fontsize=10)
ax.set_ylabel("Y")
ax.set_xlabel("X")
ax.set_zlabel("Z")
ax.set_title("3D plot of word embeddings")
```

```python
# 3D plot with vectors from the origin to each point using different colors

colors = ["blue", "red", "green", "purple", "brown", "orange"]
fig = plt.figure(figsize=(6, 10))
ax = fig.add_subplot(111, projection="3d")
for x, y, z, word, color in zip(x_coords, y_coords, z_coords, corresponding_words, colors):
    ax.quiver(0, 0, 0, x, y, z, color=color, arrow_length_ratio=0.05)
    ax.text(x, y, z, word, color=color, fontsize=10)
ax.set_ylabel("y")
ax.set_xlabel("x")
ax.set_zlabel("z")
ax.set_title("3D plot of word embeddings")
# Set the limits to prevent the arrow lines from exiting the plot
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
```

Now we can extend this mathematical operation and calculate attention weights and context vectors for each input

First we add an additional loop to calculate the dot product for all pairs of inputs




```python
# dot product of each word will result in a shape = (len(tokens), len(tokens)) - attention scores

len(corresponding_words)
```

```python
# Initializing the attention scores array

attention_scores = torch.empty(6, 6)
# Compute the attention scores

# The first loops sets the key word
for i, x_i in enumerate(inputs):
    # The second loop iterates through the whole array
    for j, x_j in enumerate(inputs):
        attention_scores[i, j] = torch.dot(x_i, x_j)
attention_scores
```

- The above approach is computationally slow and expensive o(n squared)
- To solve this using linear Algebra we can use matrix multiplication
  

```python
torch.matmul(inputs, inputs.T) == attention_scores
```

```python
# Lets use matrix multiplication to calculate attention scores

attention_scores = torch.matmul(inputs, inputs.T)
attention_scores
```

## Attention weights - Normalization of attention weights

- Using dim = -1 + softmax allows ensures normalization along the last dimension of the inputs

```python

attention_weights = torch.softmax(attention_scores, dim=-1)
attention_weights
```

```python
# Testing the attention weights summation to 1 using first row only

total = 0
for i in attention_weights:
    for j in i:
        total += j
    break
total
```

## Context Vectors

```python
# Generating the all context vector 
# Each row(mathematically but programming column) corresponds to the context vector for the input in that postition

all_context_vectors = torch.matmul(attention_weights, inputs)
all_context_vectors
```

```python
# Checking the context vector for the second input

all_context_vectors[1]
```

## Combining everything

```python
def simplifiedSimpleAttention(token_embeddings_input):
    # Get attention scores
    attention_scores = token_embeddings_input @ token_embeddings_input.T
    attention_weights = torch.softmax(attention_scores, dim=-1)
    all_context_vectors = attention_weights @ inputs
    return all_context_vectors
```

```python
all_context_vectors = simplifiedSimpleAttention(inputs)
all_context_vectors
```
